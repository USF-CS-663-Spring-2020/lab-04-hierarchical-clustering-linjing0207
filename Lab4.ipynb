## 1. The (reduced) US Census data:

```python
import pandas as pd

# The data is wrapped with this notebook.
# This is a smaller version of the 1990 census, available at UCI and other places.
# Taking a subset helps running time and to see the dendrogram.
df = pd.read_csv('USCensus1990-2000rows.data.txt')
df
```

```python
# While the implementation in sklearn is good, the scipy version can produce the dendrogram
from scipy.cluster.hierarchy import linkage, dendrogram
Z = linkage(df, 'ward')
```

```python
# Draw the dendrogram
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(12,10))
dendrogram(Z)
plt.show()
```
